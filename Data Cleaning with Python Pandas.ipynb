# Data Cleaning With Python Pandas

```python
import pandas as pd
```

### Import excel file to jupyter notebook.

```python
df = pd.read_excel(r"C:\Users\louie\Downloads\Nashville Housing Data for Data Cleaning (reuploaded) (1).xlsx")
```

### Drop duplicate rows.

```python
df = df.drop_duplicates()
df = df.dropna()
```

### Clean ParcelID.

```python
df['ParcelID']=df['ParcelID'].apply(lambda x: x[:-6])
```

### Separting Addresses and Dates to Indiviual Columns.

```python
df.loc[:,'OwnerStreet'] = df['OwnerAddress'].str.split(',').str[0]
df.loc[:,'OwnerCity'] = df['OwnerAddress'].str.split(',').str[1]
df.loc[:,'OwnerState'] = df['OwnerAddress'].str.split(',').str[2]

df.loc[:,'OwnerState']=df['OwnerState'].str.strip()

US_state = {'TN':'TENNESEE'}

df.loc[:,'OwnerState'] = df['OwnerState'].map(US_state)
```

```python
df.loc[:, 'PropertyStreet'] = df['PropertyAddress'].str.split(',').str[0]
df.loc[:, 'PropertyCity'] = df['PropertyAddress'].str.split(',').str[1]
```

```python
df.loc[:,'ParcelID'] = df['ParcelID'].str.strip(' ')
df.loc[:,'ParcelID'] = df['ParcelID'].str.replace(' ','-')
```

```python
df.loc[:,'SaleDate']= pd.to_datetime(df['SaleDate'])

df.loc[:,'SaleYear'] = df['SaleDate'].dt.year
df.loc[:,'SaleMonth'] = df['SaleDate'].dt.month



number_to_month = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}


df.loc[:,'SaleMonth'] = df['SaleMonth'].map(number_to_month)

df.loc[:, 'SaleDay'] = df['SaleDate'].dt.day

df.loc[:, 'SaleDayofWeek'] = df['SaleDate'].dt.day_name()

df.loc[:,'OwnerFirstName'] = df['OwnerName'].str.split(',').str[1]
df.loc[:,'OwnerLastName']= df['OwnerName'].str.split(',').str[0]

split_names = df['OwnerFirstName'].str.split('&', expand=True)
df.loc[:,'OwnerFirstName1'] = split_names[0]
df.loc[:,'OwnerFirstName2'] = split_names[1]

df.loc[:,'OwnerFirstName2'] = df['OwnerFirstName2'].apply(lambda x : '' if x==None else x ) 



```

```python
print(df)
```

### Drop columns.

```python
droppables = ['SaleDate', 'OwnerName', 'OwnerFirstName','OwnerAddress']


df = df.drop(columns = droppables)
```

### Change remaining columns to uppoer cases to match other columns.

```python
df[['SaleDayofWeek','SoldAsVacant']] = df[['SaleDayofWeek','SoldAsVacant']].apply(lambda x : x.str.upper())
```
